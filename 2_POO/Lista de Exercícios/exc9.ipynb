```python
class Cliente:
    
    '''
    Essa classe cria uma representação de um cliente
    '''
    
    def __init__(self, nome, idade, email):
        
        '''
        Parâmetros
        ----------
        
        nome : str
            nome do cliente
            
        idade : int
            idade do cliente
            
        email : str
            email do cliente
        '''
        
        self.nome = nome
        self.idade = idade
        self.email = email
        
    def __repr__(self):
        
        texto = "Nome: {}\nIdade: {}\nE-mail: {}".format(
            self.nome,
            self.idade,
            self.email)
        
        return texto
```
