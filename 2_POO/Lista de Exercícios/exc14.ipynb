14. Faça uma classe ContaVip que difere da ContaCorrente por ter cheque especial (novo atributo) e é filha da classe ContaCorrente. Você precisa implementar os métodos para saque, transferência ou depósito?

```python
class ContaVip(ContaCorrente):
    
    '''
    Classe que representa a conta VIP de um cliente
    
    '''
    
    def __init__(self, cliente, saldo=0, cheque_especial=0):
        
        '''
        Construtor
        
        Parâmteros
        ----------
        
        cliente : obj
            cliente que estará atrelado(a) à conta VIP
            
        saldo : int
            saldo inicial da conta VIP
            
        cheque_especial : int
            valor do cheque especial disponível para o cliente VIP
            (inicia-se com o valor zero)
        
        '''
        
        super().__init__(cliente, saldo)
        self.cheque_especial = cheque_especial
    
    def saque(self, valor):
        
        '''
        Saca dinheiro da conta VIP do cliente
        
        Parâmetros
        ----------
        
        valor : int
            valor a ser sacado
            
        '''
        
        if ((self.__saldo + self.cheque_especial) - valor >= 0):
            self.__saldo -= valor
        else:
            print("Saldo insuficiente.")
            
    def deposito(self, valor):
        
        '''
        Deposita dinheiro na conta VIP do cliente
        
        Parâmetros
        ----------
        
        valor : int
            valor a ser depositado
            
        '''
        
        self.__saldo += valor
        
    def transferencia(self, other, valor):
        
        '''
        Transfere dinheiro para a conta de um outro cliente
        
        Parâmetros
        ----------
        
        other : obj
            conta da pessoa que receberá a transferência
        
        valor : int
            valor a ser transferido
            
        '''
        
        if (self.__saldo + self.cheque_especial) - valor >= 0:
            self.__saldo -= valor
            other.__saldo += valor
        else:
            print("Saldo insuficiente.")
        
```
