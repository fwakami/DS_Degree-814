```python
class Bola:        
    
    def __init__(self, cor, raio):
        '''
        Cria uma bola
        
        Parâmetros
        ----------
        cor : str
            cor da bola
        raio : float
            raio da bola
            
        '''
        self.cor = cor
        self.raio = raio
        self.area = self.calcula_area()
        self.volume = self.calcula_volume()
        
    def imprime_cor(self):
        '''
        imprime a cor da bola
        
        '''
        print(f'A cor da bola é {self.cor}')
        
    def calcula_area(self):
        '''
        Calcula a área da bola
        
        '''
        pi = 3.14
        area = 4 * pi * self.raio * self.raio
        return area
    
    def calcula_volume(self):
        '''
        Calcula o volume da bola
        
        '''
        pi = 3.14
        volume = ((4 * pi * self.raio * self.raio * self.raio)/3)
        return volume
    
    def __repr__(self):
        return f"A bola é da cor {self.cor}, tem area de área de {self.area} e volume de {self.volume}"
    
bola_vermelha = Bola(
    cor = "vermelha",
    raio = 4)

bola_vermelha
```

```python
class retangulo:
    def __init__(self, lado_a, lado_b):
        '''
        Calcula a área do retângulo
        
        
        Parâmetros
        ----------
        
        lado_a: float
            base do retangulo
            
        lado_b: float
            altura do retangulo
            
        '''
        
        self.base = lado_a
        self.altura = lado_b
    
    def area_retangulo(self):
        self.area_retangulo = self.base * self.altura
        return self.area_retangulo
    
    def __repr__(self):
        return f"O retângulo tem {self.base * self.altura} de área."
    
retangulo_1 = retangulo(
    lado_a = 4,
    lado_b = 5)

retangulo_1
```

```python
class Funcionario:
    
    def __init__ (self, nome, email):
    
        '''
        Cria um funcionário
        
        Parâmetros
        ----------
        
        nome : str
            nome do funcionário
            
        email : str
            email do funcionário
            
        '''
        
        self.nome = nome
        self.email = email
        
        # horas trabalhadas no mês do funcionário
        self.horas_trabalhadas = {}
        
        # valor da hora naquele mês
        self.salario_hora = {}
        
    def lancar_horas_mes(self, mes, quantidade_horas):
        
        '''
        Lança a quantidade de horas trabalhadas no mês
        
        Parâmetros
        ----------
        
        mês : str
            mês vigente, exemplo: "10/2021"
            
        quantidade_horas : int
            quantidade de horas trabalhadas no mês
            
        '''
        
        self.horas_trabalhadas[mes] = quantidade_horas
        
    def define_salario_hora_mes(self, mes, valor_salario_hora):
        
        '''
        Define o valor do salário por hora trabalhada naquele mês
        
        Parâmetros
        ----------
        
        mês : str
            mês vigente, exemplo: "10/2021"
            
        quantidade_horas : float
            valor da hora no mês vigente
            
        '''
        
        self.salario_hora[mes] = valor_salario_hora
        
    def calcula_salario_mes(self, mes):
        
        '''
        Calcula o salário para o mês vigente
        
        Parâmetros
        ----------
        
        mês : str
            mês vigente, exemplo: "10/2021"
            
        Retorno:
        --------
        
        out : float
            valor do salário no mês vigente
            Retorna 0 se não trabalhou no mês,
            ou se o mês não está cadastrado.
            
        '''
        
        if ((mes in self.salario_hora) and (mes in self.horas_trabalhadas)):
        
            salario_mes = self.salario_hora[mes] * self.horas_trabalhadas[mes]
        
            return salario_mes
            
        print('Salário ou hora não cadastrada.')
        return 0
    
    def __repr__(self):
        return f"O nome do funcionário é {self.nome} e seu e-mail é '{self.email}'."
```

```python
class Televisor:
    
    '''
    Cria uma representação de um televisor
        
    '''
    
    def __init__(self, fabricante, modelo):
        
        '''
        Construtor
        
        Parâmetros
        ----------
        
        fabricante : str
            fabricante do televisor
            
        modelo : str
            modelo do televisor
            
        canal : int
            canal sintonizado ou a ser adicionado na lista de canais
            
        volume : int
            volume atual configurado no televisor
            
        lista_de_canais : list [int]
            lista de canais salvos ao televisor
        
        '''
        
        self.fabricante = fabricante
        self.modelo = modelo
        self.canal = None
        self.lista_de_canais = []
        self.volume = 50
        
    def aumentar_volume(self):
        
        '''
        Este método aumenta o volume da TV em 5 unidades, 
        considerando que o volume padrão incial seja 50
        
        Parâmetros
        ----------
        
        volume : int
            volume da TV
            
        '''
        
        if self.volume >= 0 and self.volume < 100:
            self.volume = self.volume + 5
            print(f"O volume da TV está em {self.volume}")
            
        else:
            print("O volume da TV já está no máximo.")
        
    def diminuir_volume(self):
        
        '''
        Este método diminui o volume da TV em 5 unidades, 
        considerando que o volume padrão incial seja 50
        
        Parâmetros
        ----------
        
        volume : int
            volume da TV
            
        '''
        
        if self.volume > 0 and self.volume <= 100:
            self.volume = self.volume - 5
            print(f"O volume da TV está em {self.volume}")
            
        else:
            print("O volume da TV já está no mínimo.")
            
    def sintonizar_novo_canal(self, canal, lista_de_canais=[]):
        
        '''
        Este método adiciona um novo canal na sua lista de canais
        
        Parâmetros
        ----------
        
        canal : int
            canal a ser inserido na lista de canais
        
        lista_de_canais : list[int]
            lista de canais disponíveis na TV
            
        '''
        
        if canal not in self.lista_de_canais:
            self.lista_de_canais.append(canal)
            print(f"O canal {canal} foi adicionado à sua lista de canais.")
            
        else:
            print("Este canal já está na sua lista de canais.")
            
    def trocar_canal(self, canal):
        
        '''
        Este método sintoniza o canal desejado para ser assistido na TV
        
        Parâmetros
        ----------
        
        canal : int
            canal desejado a ser assistido na TV
            
        '''
        
        if canal in self.lista_de_canais:
            print(f"O canal {canal} foi sintonizado.")
            
        else:
            print(f"O canal desejado não encontra-se na sua lista de canais disponíveis.")
            
    def __repr__(self):
        return f"Esta televisão é da {self.fabricante} e seu modelo é o {self.modelo}."
```

```python
import time

class Cronômetro:
    
    '''
    Essa classe simula um cronômetro
    
    '''
    
    def __init__(self, horas = 0, minutos = 0, segundos = 0):
        
        '''
        
        Construtor
        
        Parâmetros
        ----------
        
        horas : int
            Indicador de horas do cronômetro
            
        minutos : int
            Indicador de minutos do cronômetro
            
        segundos : int
            Indicador de segundos do cronômetro
        
        '''
        
        self.horas = horas
        self.minutos = minutos
        self.segundos = segundos
        
    def __repr__(self):
        
        '''
        Display do cronômetro a ser mostrado em tela
        
        '''
        
        contagem = '%.2d:%.2d:%.2d' %(self.horas, self.minutos, self.segundos)
        return contagem
    
    def funcionamento_cronometro (self):
        
        if self.segundos >= 60:
            self.segundos = 0
            self.minutos += 1
            
        if self.minutos >= 60:
            self.minutos = 0
            self.horas += 1
            
        self.segundos += 1
        
    def startar_cronômetro (self):
        
        i = 0
        
        while i == 0:
            
            print(self)
            self.funcionamento_cronometro()
            time.sleep(1)
            
```
